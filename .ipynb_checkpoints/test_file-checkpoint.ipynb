### Import Libraries

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow.keras.preprocessing.image import load_img,img_to_array
from tensorflow.keras.models import load_model
```

### Model Loading

```python
model=load_model("models/DenseNet201_model.h5")
```

```python
class_labels={0:'Apple___Apple_scab', 
              1:'Apple___Black_rot', 
              2:'Apple___Cedar_apple_rust', 
              3:'Apple___healthy'}
```

### User input

```python
filepath="input/val/healthy/healthy (12).JPG"
```

```python
image=load_img(filepath, grayscale=False, color_mode='rgb', target_size=(256,256))
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
image=img_to_array(image)
image=image/255.0
expanded_image=np.expand_dims(image,axis=0)
```

```python
model_prediction=model.predict(expanded_image,verbose=1)
print(model_prediction)
```

```python
print(f"model predicted class is -> {np.argmax(model_prediction[0])}")
print(f"model predicted label is -> {class_labels[np.argmax(model_prediction[0])]}")
```
